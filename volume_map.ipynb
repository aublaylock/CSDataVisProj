```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import plotly

noise_complaints = pd.read_csv("311_Service_Requests_from_2010_to_Present_20251209.csv")
```

```python
locations = noise_complaints[["Longitude", "Latitude"]].dropna()
locations = locations[pd.to_numeric(locations['Longitude'], errors='coerce').notnull()]
locations = locations[pd.to_numeric(locations['Latitude'], errors='coerce').notnull()]
```

```python
locations_sample = locations[0:500000]
locations_with_rounding = locations_sample.copy()
locations_with_rounding['Latitude Rounded'] = locations_with_rounding['Latitude'].round(2)
locations_with_rounding['Longitude Rounded'] = locations_with_rounding['Longitude'].round(2)
```

```python
bins = locations_with_rounding.groupby(['Latitude Rounded','Longitude Rounded']).size().reset_index(name='count')
bins[['count']] = bins[['count']]*5
bins.head()
```

```python
import plotly.express as px
fig = px.scatter_map(
    bins, 
    lat='Latitude Rounded', 
    lon='Longitude Rounded', 
    hover_name='count', 
    hover_data={'count':False, "Latitude Rounded":False, "Longitude Rounded":False}, 
    zoom=10, 
    height=600, 
    map_style='open-street-map'
)
fig.update_layout(margin={'r':0,'t':0,'l':0,'b':0})
fig
```
