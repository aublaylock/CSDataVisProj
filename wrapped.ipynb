```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly

noise_complaints = pd.read_csv("311_Service_Requests_from_2010_to_Present_20251209.csv")
```

```python
noise_complaints['Descriptor'].value_counts()
```

```python
noise_complaints = noise_complaints[
    (noise_complaints.Descriptor == 'Loud Music/Party') | 
    (noise_complaints.Descriptor == 'Banging/Pounding') |
    (noise_complaints.Descriptor == 'Loud Talking') |
    (noise_complaints.Descriptor == 'Car/Truck Music') |
    (noise_complaints.Descriptor == 'Noise: Construction Before/After Hours (NM1)') |
    (noise_complaints.Descriptor == 'Car/Truck Horn') |
    (noise_complaints.Descriptor == 'Engine Idling') |
    (noise_complaints.Descriptor == 'Noise, Barking Dog (NR5)') |
    (noise_complaints.Descriptor == 'Noise: Construction Equipment (NC1)') |
    (noise_complaints.Descriptor == 'Loud Television')
    ]
```

```python
manhattan_complaints = noise_complaints[noise_complaints.Borough == 'MANHATTAN']
manhattan_type_counts = dict(manhattan_complaints['Descriptor'].value_counts())

labels = []
sizes = []
for x, y in manhattan_type_counts.items():
    labels.append(x)
    sizes.append(y)
colors = sns.color_palette('bright')
plt.pie(sizes, labels=labels, colors=colors, autopct='%.0f%%')
plt.show()
```
