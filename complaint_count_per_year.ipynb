```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import plotly

noise_complaints = pd.read_csv("311_Service_Requests_from_2010_to_Present_20251209.csv")
```

```python
dates = noise_complaints.copy()[['Created Date']]
dates['year'] = dates['Created Date'].str[6:10].astype(int)
dates['year'].unique()
```

```python
year = dates.groupby('year').agg(complaint_count=('year', 'count'))
year
```

```python
sns.set_theme(rc={'figure.figsize':(10,6)})
plot = sns.barplot(data=year, x='year', y='complaint_count')
plot.bar_label(plot.containers[0], fontsize=10)
```

```python
dates['month'] = dates['Created Date'].str[0:2].astype(int)
dates['month'].unique()
```

```python
month = dates.groupby('month').agg(complaint_count=('month', 'count'))
month
```

```python
sns.set_theme(rc={'figure.figsize':(10,6)})
plot = sns.barplot(data=month, x='month', y='complaint_count')
plot.bar_label(plot.containers[0], fontsize=10)
```
